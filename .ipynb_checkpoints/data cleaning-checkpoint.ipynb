# Movie Data Wrangling

```python
import numpy as np
import pandas as pd
```

```python
origin_data = pd.read_csv('movies.csv')
```

```python
origin_data.describe()
```

```python
origin_data.info()
```

```python
origin_data.head()
```

```python
print len(origin_data)
print len(origin_data['id'].unique())
```

```python
drop_repeated_data = origin_data.drop_duplicates(keep="first", inplace=False)
```

```python
print len(drop_repeated_data)
print len(drop_repeated_data['id'].unique())
```

### movie type

```python
needed_data = drop_repeated_data[['id','vote_average', 'popularity','genres', 'release_year']]
needed_data.head()
```

```python
needed_data.info()
```

```python
cleaned_data = needed_data.dropna()
cleaned_data.head()
```

```python
splited_genres = cleaned_data['genres'].str.split(pat='|', expand=True)
splited_genres.head()
```

```python
splited_genres['id'] = cleaned_data['id']
splited_genres.head()
```

```python
added_data = cleaned_data.merge(splited_genres)
added_data.head()
```

```python
melted_data = pd.melt(
    added_data, id_vars=['id','vote_average', 'popularity', 'release_year'], 
    value_vars=[0, 1, 2, 3],
    value_name='genres').drop('variable',axis=1).dropna()
melted_data.head()
```

```python
# melted_data.to_csv('genres_time.csv', index=False)
```

### 电影公司数据清理

```python
drop_repeated_data.head()
```

```python
needed_data_2 = drop_repeated_data[["id", "popularity", "original_title", "production_companies", 
                                    "release_year", "budget_adj", "revenue_adj"]]
needed_data_2.head()
```

```python
splited_companies = needed_data_2["production_companies"].str.split("|", expand=True)
splited_companies["id"] = needed_data_2["id"]
splited_companies.head()
```

```python
added_data_2 = needed_data_2.merge(splited_companies)
added_data_2.head()
```

```python
melt_data_2 = pd.melt(added_data_2, id_vars=["id", "popularity", "original_title", "release_year", 
                                             "budget_adj", "revenue_adj"],
                      value_vars=[0, 1, 2, 3, 4], 
                      value_name="production_company").drop("variable", axis=1).dropna()
melt_data_2.head()
```

```python
selected_certain_companies = melt_data_2.loc[melt_data_2['production_company'].isin(["Universal Pictures", "Paramount Pictures"])]
selected_certain_companies.info()
```

```python
selected_certain_companies_cleaned = selected_certain_companies.loc[(selected_certain_companies['budget_adj']!=0) & (selected_certain_companies['revenue_adj']!=0)]
selected_certain_companies_cleaned.info()
```

```python
selected_certain_companies_cleaned.to_csv("selected_certain_companies_3.csv", index=False)
```

### 是否是小说改编

```python
drop_repeated_data.head()
```

```python
need_data_3 = drop_repeated_data[["id", "popularity", "director", "keywords", "vote_average",
                                  "vote_count", "release_year", "budget_adj", "revenue_adj"]]
need_data_3.head()
```

```python
splited_data_3 =need_data_3["keywords"].str.split("|")
splited_data_3.head()
```

```python
splited_data_pd = pd.DataFrame(splited_data_3)
splited_data_pd.head()
```

```python
need_data_3.loc[:, 'keywords'] = splited_data_pd["keywords"]
need_data_3.head()
```

```python
need_data_3.info()
```

```python
cleaned_data_2 = need_data_3.dropna()
cleaned_data_2.info()
```

```python
cleaned_data_3 = cleaned_data_2.loc[(cleaned_data_2['budget_adj']!=0) & (cleaned_data_2['revenue_adj']!=0)]
cleaned_data_3.info()
```

```python
cleaned_data_3["keywords"].head()
```

```python
def novel(a):
    if "based on novel" in a:
        return True
    else:
        return False
```

```python
judge_novel =cleaned_data_3["keywords"].apply(novel)
judge_novel.head()
```

```python
added_data_3 = cleaned_data_3.join(judge_novel, rsuffix=" whether include 'based on novel'")
added_data_3.head()
```

```python
added_data_3.to_csv("compare the moveies based on novel with original movies_2.csv", index=False)
```

### 演员对电影的影响

```python
drop_repeated_data.head()
```

```python
needed_data_4 = drop_repeated_data[['id', 'popularity', 'original_title', 'cast', 'vote_average', 'budget_adj', 'revenue_adj']]
needed_data_4.head()
```

```python
needed_data_4.info()
```

```python
clean1 = needed_data_4.dropna()
clean1.info()
```

```python
splited = clean1['cast'].str.split('|', expand=True)
splited.head()
```

```python
splited['id'] = clean1['id']
add1 = clean1.merge(splited)
add1.head()
```

```python
melt1 = pd.melt(add1, id_vars=['id', 'popularity', 'original_title', 'vote_average', 'budget_adj', 'revenue_adj'],
                value_vars=[0, 1, 2, 3, 4],
                value_name='cast').drop('variable', axis=1).dropna()
melt1.head()
```

```python
melt1.info()
```

```python
clean2 = melt1.loc[(melt1['budget_adj']!=0) & (melt1['revenue_adj']!=0)]
clean2.info()
```

```python
clean2.head()
```

```python
clean2.to_csv("cast_moives.csv", index=False)
```
